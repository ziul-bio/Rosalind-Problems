### Problem 2 - Transcribing DNA into RNA

Given a DNA string t corresponding to a coding strand, its transcribed RNA string u is formed by replacing all occurrences of 'T' in t with 'U' in u.

*  Given: A DNA string t having length at most 1000 nt. 
* Return: The transcribed RNA string of t.

```python
coding_strand = 'GATGGAACTTGACTACGTAAATT'

def Transcrition(coding_strand):
    '''Transcribing DNA into RNA Replacing T with U'''
    return coding_strand.replace('T', 'U') 

Transcrition(coding_strand)
```

## Problem 3 - Complementing a strand of DNA

In DNA strings, symbols 'A' and 'T' are complements of each other, as are 'C' and 'G'. 
The reverse complement of a DNA string s is the string sc formed by reversing the symbols of s, then taking the complement of each symbol (e.g., the reverse complement of "GTCA" is "TGAC").

* Given: A DNA string s of length at most 1000 bp.
* Return: The reverse complement sc of s.

As we saw right above, a complementay sequence of DNA('5 to 3') is the anti sense sequence or the tamplate sequence(3' to 5').  
In other hand the, the  Reverse Complement converts a DNA sequence into its reverse, complement, or reverse-complement counterpart.  
You may want to work with the reverse-complement of a sequence if it contains an ORF on the reverse strand.

```python
Reversed_Complement_Nucleotides = {'A': 'T', 'T': 'A', 'G': 'C', 'C': 'G'}
seq = 'AAAACCCGGT'

def reverse_complement(DNA_seq):
    '''Reverse Complement converts a DNA sequence into its reverse, of its sequence complement
    In other words, takes the complementary sequence, with As and Gs replaced with Ts and Cs, and return the reverse of it'''
    return ''.join([Reversed_Complement_Nucleotides[nuc] for nuc in DNA_seq[::-1]])

# Testing the function:
reverse_complement(seq)
```

## Problem 5 - Computing GC Content
The GC-content of a DNA string is given by the percentage of symbols in the string that are 'C' or 'G'. For example, the GC-content of "AGCTATAG" is 37.5%. Note that the reverse complement of any DNA string has the same GC-content.

DNA strings must be labeled when they are consolidated into a database. A commonly used method of string labeling is called FASTA format. In this format, the string is introduced by a line that begins with '>', followed by some labeling information. Subsequent lines contain the string itself; the first line to begin with '>' indicates the label of the next string.

In Rosalind's implementation, a string in FASTA format will be labeled by the ID "Rosalind_xxxx", where "xxxx" denotes a four-digit code between 0000 and 9999.

Given: At most 10 DNA strings in FASTA format (of length at most 1 kbp each).

Return: The ID of the string having the highest GC-content, followed by the GC-content of that string. Rosalind allows for a default error of 0.001 in all decimal answers unless otherwise stated; please see the note on absolute error below.

```python
# Firt at all, I need to creat a function to count the GC content
seq = 'CCACCCTCGTGGTATGGCTAGGCATTCAGGAACCGGAGAACGCTTCAGACCAGCCCGGACTGGGAACCTGCGGGCAGTAGGTGGAAT'
def gc_content(seq):
    """GC Content in a DNA or RNA sequence is the counts of Gs and Cs / by the len of the sequence"""
    return round((seq.count('C') + seq.count('G')) / len(seq) * 100, 4) 

gc_content(seq)
```

```python
# Now I have to open a Fasta_sample file:
with open('files/rosalind_gc.txt', 'r') as fasta:
    fasta_file = fasta.readlines()
    fasta_file =[l.strip() for l in fasta_file]
    print('This is the file the way we open it, we need to clean this up, joining the sequences:')
    print(fasta_file)
    print()

# Organizing the date into a dictionary, which we can acess by key:
fasta_dict = {} # Empty dictionary
tmp_label = ''  # Temporary label

for line in fasta_file:
    ''' Here we loop over the file, line by line, if the line starts with >, then, it will be a new key at fasta_dict
    and this key will have a value of a empty string.
    As the loop goes on, if a line doesn't starts with >, it'll be attributed as a value to tmp_label.
    If a new line with > is found, it'll starts over a new key'''
    if '>' in line:
        tmp_label = line
        fasta_dict[tmp_label] = ''
    else:
        fasta_dict[tmp_label] += line
print('Now with a dictionary, we can access the values by key and values')
print(fasta_dict)
```

```python
# Now with a dictionary conprehension, we can apply the gc_conten() function to all values of the dictionary.
# It'll retur a dict with the GC contents
result = {key:gc_content(value) for key, value in fasta_dict.items()}
print(result)
```

```python
# Printing the anwser to problem 5:
# Now it's the easiest part, we just need to return the max value.
max_GC = max(result, key=result.get)
print('The biggest values of GCs content in the file is:',max_GC.strip('>'), result[max_GC])
```
